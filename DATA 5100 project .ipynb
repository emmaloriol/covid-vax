```python
fd = pd.read_csv("https://raw.githubusercontent.com/emmaloriol/covid-vax/main/DECENNIALPL2020.csv")
fd
```

```python
pop = pd.DataFrame(fd.iloc[0,2:])
#pop['State'] = pop['0']
pop.reset_index(inplace = True)
pop['State'] = pop['index']
pop.drop('index',axis = 1,inplace= True)
pop['population'] = pop[0]
pop.drop(0,axis = 1,inplace= True)
pop
```

```python
result5['Population'] = pop['population']

result5['Population'] = [''.join(str(i).split(',')) for i in result5['Population']]

result5['Population'] = result5['Population'].astype(int)

result5['cases per capita'] = result5['sum_of_tot_cases']/result5['Population']

result5['Distributed'] = result['Distributed']
result5['Vaccine Coverage'] = result5['Distributed'] / result5['Population']
result5
```

```python
result8 = result5.sort_values(by="cases per capita",ascending=False)
```

```python
result8[['cases per capita','State']].plot(kind='bar',x='State',figsize=(19,12))
```

```python
sns.set(rc={"font.style":"normal",
            "text.color":"black",
            'axes.titlesize':18,
            "xtick.color":"black",
            "ytick.color":"black",
            "axes.grid":True,
            'axes.labelsize':30,
            'xtick.labelsize':25,
            'font.size':20,
            'figure.figsize':(16.7,9.27),
            'ytick.labelsize':20})
sns.barplot(data = result8.head(), x = 'State', y = 'cases per capita', palette = "bright", alpha = 0.6)

plt.title("Top 5 States with Highest cases per capita")
plt.xlabel('States In USA', fontsize = 18)
plt.ylabel('Cases per capita' ,fontsize = 18)

plt.tick_params(labelsize = 15);
```
